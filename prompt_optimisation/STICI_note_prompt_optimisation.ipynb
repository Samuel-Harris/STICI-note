```python
import chromadb
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=100,
    length_function=len,
    is_separator_regex=False,
)
```

```python
db_dir = "./chroma_db"
collection_prefix = "document_"
chroma_client = chromadb.PersistentClient(path=db_dir)
```

```python
embedding_function = SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
```

```python
collection_names: dict[int, str] = {}
for i, document in documents_df[["index", "text"]].iterrows():
    collection_name = f"{collection_prefix}{i}"
    collection_names[i] = collection_name
    print(f"Creating collection {collection_name}")

    chunks = [chunk.page_content for chunk in text_splitter.create_documents([document["text"]])]

    try:
        chroma_client.delete_collection(name=collection_name)
    except ValueError:
        pass

    collection = chroma_client.create_collection(collection_name,
                                                 embedding_function=embedding_function)
    collection.add(
        documents=chunks,
        ids=[str(i) for i in range(len(chunks))]
    )
```

```python
retrievers = {
    i: ChromadbRM(
        collection_names[i],
        db_dir,
        embedding_function=embedding_function,
        k=3
    ) for i in collection_names
}
```

```python
no_answer_response = "The answer to your question is not in the provided document."
def generate_example(row, documents):
    return Example(question=row[1]["question"],
                   answer=row[1]["answer"] if "answer" in row[1] else no_answer_response,
                   document_index=row[1]["document_index"]).with_inputs("question", "document_index")
```

```python
no_answer_questions = list(map(lambda row: generate_example(row, documents_df), no_answer_questions_df.iterrows()))
single_passage_questions = list(map(lambda row: generate_example(row, documents_df), single_passage_questions_df.iterrows()))
multi_passage_questions = list(map(lambda row: generate_example(row, documents_df), multi_passage_questions_df.iterrows()))
validate_questions = no_answer_questions + single_passage_questions + multi_passage_questions
```

```python
class GenerateAnswer(dspy.Signature):
    """Answer questions with short factoid answers."""

    context = dspy.InputField(desc="may contain relevant facts")
    question = dspy.InputField()
    answer = dspy.OutputField(desc=f"If there is an answer, it will consist of information from the context. If the answer can not be found from the context, a response of \"{no_answer_response}\" should be given.")
```

```python
class RAG(dspy.Module):
    def __init__(self, num_passages=3):
        super().__init__()

        self.generate_answer = dspy.Predict(GenerateAnswer)

    def retrieve(self, question, document_index):
        return [passage["long_text"] for passage in retrievers[document_index].forward(query_or_queries=[question])]

    def forward(self, question, document_index):
        context = self.retrieve(question, document_index)
        prediction = self.generate_answer(context=context, question=question)
        return dspy.Prediction(context=context, answer=prediction.answer)
```

```python
train_validate_answer_embeddings = {example.question: embedding_function([example.answer]) for example in validate_questions}
```

```python
def semantic_similarity(answer_embeddings) -> float:
    def similarity_function(example: Example, pred: Prediction, trace=None) -> float:
        pred_embedding = embedding_function([pred.answer])
        actual_embedding = answer_embeddings[example.question]
        return cosine_similarity(actual_embedding, pred_embedding)[0][0]
    return similarity_function
```

```python
# Set up a basic teleprompter, which will compile our RAG program.
teleprompter = BootstrapFewShotWithRandomSearch(
    metric=semantic_similarity(train_validate_answer_embeddings),
    # teacher_settings={"lm": llama_3_1},
)
```

```python
from transformers import AutoModelForCausalLM, BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(load_in_4bit=True)
```

```python
import gc
```

```python
for model_name, compile_path in models.items():
    print(f"Loading {model_name}")
    model = dspy.HFModel(model=model_name,
                         model_kwargs={"quantization_config": quantization_config})
    dspy.settings.configure(lm=model)

    print(f"Compiling {model_name}")
    compiled_rag = teleprompter.compile(RAG(), trainset=validate_questions)
    compiled_rag.save(compile_path)

    print(f"Removing {model_name} from memory")
    del model
    gc.collect()
    torch.cuda.empty_cache()
```
